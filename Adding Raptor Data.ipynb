```python
import sys
import csv

csv.field_size_limit(sys.maxsize)

def readYearRecords():
    with open('yearRecords.csv', mode='r') as infile:
        reader = csv.reader(infile)
        with open('yearReader.csv', mode='w') as outfile:
            writer = csv.writer(outfile)
            mydict = {rows[0]:rows[1] for rows in reader}

    for name in mydict:
        val = mydict[name]
        mydict[name] = eval(val)

    return mydict




def readRaptor():
    with open('historical_RAPTOR_by_player.csv', mode='r') as infile:
        reader = csv.reader(infile)
        with open('raptorReader.csv', mode='w') as outfile:
            writer = csv.writer(outfile)
            mydict = {rows[0]:rows[1] for rows in reader}

    reader = csv.DictReader(open('historical_RAPTOR_by_player.csv'))

    result = {}
    for row in reader:
        playerId = row['player_id']
        season = row['season']
        if playerId in result:
            result[playerId][season] = row
        else:
            result[playerId] = {season:row}

    return result

def writeData(yearRecords):
    w = csv.writer(open("rapYearRecords.csv", "w"))
    for key, val in yearRecords.items():
        w.writerow([key, val])
```

```python
def getPlayerId(playerName):
    if playerName not in playersToIds:
        sp = playerName.rindex(' ')
        truncName = playerName[:sp]
        return playersToIds[truncName]
    else:
        return playersToIds[playerName]

def appendRaptor(yearRecords, raptorData):
    newRapData = {}

    for year in yearRecords:
        if year == "2020":
            continue
        entry = yearRecords[year]
        for teams in entry:
            for playerName, stats in teams[0].items():
                playerId = getPlayerId(playerName)
                if playerId not in raptorData:
                    print(playerName)
                playerRapData = raptorData[playerId]
                tilNow, averaged = averageRaps(playerRapData, year)
                if len(stats) == 64:
                    stats += tilNow
                    stats += averaged
                    
    
                
def averageRaps(playerRapData, year):
    tilNow = [0]*3
    count = 0
    for rapYear, rapEntry in playerRapData.items():
        if int(rapYear) <= int(year):
            yearData = list(rapEntry.values())[5:8]
            for j in range(len(yearData)):
                tilNow[j] += float(yearData[j])
            count += 1
    averaged = [x / count for x in tilNow.copy()] if count != 0 else tilNow
#     averaged = tilNow.copy() / count if count != 0 else tilNow
    return tilNow, averaged
```
