```python
url = "https://www.basketball-reference.com/teams/MIL/2020.html"

html = urlopen(url)
soup = BeautifulSoup(html)

players = soup.find('tbody').findAll('tr')
pSet = set()
sharedTimes = {}
for player in players:
    playerName = player.find('a').text
    if "III" in playerName:
        ind = playerName.rindex(" ")
        playerName = playerName[:ind]
    pSet.add(playerName)
    sharedTimes[playerName] = NBAPlayerAnalysis.findTotalSharedTimes(playerName)

bucksSharedTimes = sharedTimes
%store bucksSharedTimes
%store pSet
sharedTimes
```

```python

totalTimes = {}

for player in bucksSharedTimes:
    print(player)
    for teammate in bucksSharedTimes[player]:
        if teammate in pSet:
            pairing1 = (player, teammate)
            pairing2 = (teammate, player)
            if pairing1 in totalTimes:
                totalTimes[pairing1] = max(bucksSharedTimes[player][teammate], totalTimes[pairing1])
            elif pairing2 in totalTimes:
                totalTimes[pairing2] = max(bucksSharedTimes[player][teammate], totalTimes[pairing2])
            else:
                totalTimes[pairing1] = bucksSharedTimes[player][teammate]
totalTimes
```

```python
pStreaks = {}
for player in pSet:
    streaks = NBAPlayerAnalysis.findStreaks(player)
#     print(player, streaks)
    pStreaks[player] = streaks
pStreaks
```

```python
sharedSeasons = 0
streaksOverlap = 0
daysOverlap = 0
for pairing in totalTimes:
    time = totalTimes[pairing]
    
    if time >= datetime.timedelta(days = 0):
        player1 = pairing[0]
        player2 = pairing[1]

        p1teams = NBAPlayerAnalysis.getTeams(player1)
        p2teams = NBAPlayerAnalysis.getTeams(player2)

        for year in p1teams:
            if year in p2teams:
                if p1teams[year][0] == p2teams[year][0]:
                    print(pairing)
                    print(year)
                    print(p1teams[year][0])
                    sharedSeasons += 1
                    
                    p1streaks = pStreaks[player1][0][year]
                    p2streaks = pStreaks[player2][0][year]
                    
                    for s in p1streaks:
                        for t in p2streaks:
                            
                            s0 = datetime.datetime.strptime(s[0], '%Y-%m-%d')
                            s1 = datetime.datetime.strptime(s[1], '%Y-%m-%d')
                            t0 = datetime.datetime.strptime(t[0], '%Y-%m-%d')
                            t1 = datetime.datetime.strptime(t[1], '%Y-%m-%d')
                            

                            if (s0 >= t0 and s0 <= t1):
                                if s1 <= t1: # s is fully inside t
                                    print("s:", s)
                                    print("t:", t)
                                    daysOverlap += (s1-s0).days + 1
                                    print('1:', (s1 - s0).days + 1)
                                else: # s isn't fully inside t
                                    print("s:", s)
                                    print("t:", t)
                                    daysOverlap += (t1 - s0).days + 1
                                    print('2:', (t1 - s0).days + 1)
                                streaksOverlap +=1
                            elif (s1 >= t0 and s1 <= t1):
                                
                                print("s:", s)
                                print("t:", t)
                                daysOverlap += (s1 - t0).days + 1
                                print('3:', (s1 - t0).days + 1)
                                streaksOverlap +=1
                            elif (s0 <= t0 and s1 >= t1): # t is fully inside s
                                print("s:", s)
                                print("t:", t)
                                daysOverlap += (t1 - t0).days + 1
                                print('4:', (t1 - t0).days + 1)
                                streaksOverlap +=1
print(sharedSeasons)
print("st", streaksOverlap)
#         print(pairing)
#         print(p1teams)
```
