```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
import pandas as pd
from selenium import webdriver

import string
import datetime
```

```python
def getTeamMap():
    url = "https://stats.nba.com/teams/"
    options = setupHeadlessChrome()

    driver = webdriver.Chrome(options=options) # used to view generated JavaScript
    driver.get(url)

    html = driver.page_source
    teamToId = {}
    teams = driver.find_elements_by_class_name("stats-team-list__link")
    for team in teams:
        teamStr = team.get_attribute('href')
        teamId = parseTeamId(teamStr)
        
        abbr = team.find_element_by_tag_name('img').get_attribute('abbr')
        teamToId[abbr] = teamId

        teamName = team.get_attribute('innerText')
        teamName = teamName.lstrip()
        teamName = teamName.rstrip()
        teamToId[teamName] = teamId
        
        spaceInd = teamName.rindex(' ')
        teamName = teamName[spaceInd+1:]
        teamToId[teamName] = teamId
    return teamToId

def parseTeamId(teamStr):
    teamInd = teamStr.index('team/')
    teamId = teamStr[teamInd+5:len(teamStr) - 1]
    return teamId

def doExceptions(NBAMap):
    NBAMap['BRK'] = NBAMap['BKN']
    NBAMap['PHO'] = NBAMap['PHX']
    NBAMap['NOH'] = NBAMap['NOP']
    NBAMap['NOK'] = NBAMap['NOP']
    NBAMap['NJN'] = NBAMap['BKN']
    NBAMap['CHO'] = NBAMap['CHA']
    NBAMap['Bobcats'] =  NBAMap['Hornets']
    NBAMap['Sonics'] = NBAMap['Oklahoma City Thunder']
    NBAMap['SEA'] = NBAMap['OKC']
    return NBAMap

def setupHeadlessChrome():
    options = webdriver.ChromeOptions() # don't open URL window
    options.add_argument('--headless')
    prefs = {"profile.managed_default_content_settings.images": 2, 'disk-cache-size': 4096 }
    options.add_experimental_option("prefs", prefs)
    return options
```

```python
statsNBAMap = getTeamMap()
statsNBAMap = doExceptions(statsNBAMap)
%store statsNBAMap
```
